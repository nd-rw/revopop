```python
import pandas as pd
import sqlite3
from datetime import datetime, timedelta
```

```python
conn = sqlite3.connect('../revopop.db')
query = "SELECT * FROM gym_population;"
```

```python
gym_pop = pd.read_sql_query(query,conn)
```

```python
gym_pop.tail()
```

```python
# add 8 hours (28800 seconds) to convert to Perth time

gym_pop['unix_time_wa'] = gym_pop['unix_time'] + 28800
```

```python
gym_pop.tail()
```

```python
gym_pop['date'] = pd.to_datetime(gym_pop['unix_time_wa'],unit='s')
```
